```python
import os
import re

# ==========================
# Folder paths
# ==========================

input_folder = r"C:\Users\bisho\OneDrive\Desktop\task 0\raw books"
output_folder = r"C:\Users\bisho\OneDrive\Desktop\task 0\cleaned_books"

os.makedirs(output_folder, exist_ok=True)

# ==========================
# Process all books
# ==========================

for filename in os.listdir(input_folder):

    if filename.lower().endswith(".txt"):

        print(f"Processing: {filename}")

        input_path = os.path.join(input_folder, filename)

        with open(input_path, "r", encoding="utf-8", errors="ignore") as f:
            text = f.read()

        # Remove Project Gutenberg header
        start = re.search(
            r"\*\*\* START OF .*? \*\*\*",
            text,
            flags=re.IGNORECASE | re.DOTALL
        )

        if start:
            text = text[start.end():]

        # Remove Project Gutenberg footer
        end = re.search(
            r"\*\*\* END OF .*? \*\*\*",
            text,
            flags=re.IGNORECASE | re.DOTALL
        )

        if end:
            text = text[:end.start()]

        # Basic cleaning
        text = re.sub(r'[ \t]+', ' ', text)
        text = re.sub(r'\n{3,}', '\n\n', text)
        text = text.strip()

        # Save cleaned book
        output_name = filename.replace(".txt", "_clean.txt")
        output_path = os.path.join(output_folder, output_name)

        with open(output_path, "w", encoding="utf-8") as f:
            f.write(text)

        print(f"Saved -> {output_name}")

print("\nAll books cleaned successfully!")
```

```python
paragraphs = re.split(r"\n\s*\n", book_text)

paragraphs = [
    p.strip()
    for p in paragraphs
    if 100 <= len(p.split()) <= 200
]

print("Paragraphs:", len(paragraphs))
```

```python
topics = {

    "Crime and Justice":[
        "crime","justice","law","judge","court",
        "police","criminal","trial","prison"
    ],

    "Scientific Discovery and Technology":[
        "science","technology","machine","experiment",
        "laboratory","engine","invent","discovery"
    ],

    "Human Nature and Morality":[
        "moral","virtue","vice","good",
        "evil","character","honest","conscience"
    ],

    "Class, Society, and Social Evolution":[
        "society","class","rich","poor",
        "family","education","social","wealth"
    ],

    "Imperialism and Colonialism":[
        "empire","colonial","colony",
        "government","rule","nation","power"
    ]
}
topic_data = {}

for topic in topics:
    topic_data[topic] = []

for para in paragraphs:

    para_lower = para.lower()

    best_topic = None
    best_score = 0

    for topic, words in topics.items():

        score = 0

        for word in words:
            score += para_lower.count(word)

        if score > best_score:
            best_score = score
            best_topic = topic

    if best_topic is not None and best_score > 0:
        topic_data[best_topic].append(para)
for topic in topic_data:
    print(topic)
    print(len(topic_data[topic]))
    print()
```

```python
# pip install -U google-genai pandas pydantic

import os
import time
import pandas as pd
from pydantic import BaseModel, Field
from google import genai
from google.genai import types
from getpass import getpass

# =====================================
# Configuration & Secure Authentication
# =====================================

API_KEY = getpass("Enter your Gemini API Key (starts with AIzaSy): ")
client = genai.Client(api_key=API_KEY)

MODEL = "gemini-3.1-flash-lite"

# Tentukan penulis yang ingin Anda tiru di sini
TARGET_AUTHOR = "Charles Dickens" 

topics = [
    "Crime and Justice",
    "Scientific Discovery and Technology",
    "Human Nature and Morality",
    "Class, Society, and Social Evolution",
    "Imperialism and Colonialism",
]

PARAGRAPHS_PER_TOPIC = 50
BATCH_SIZE = 2  # Diturunkan ke 2 demi keamanan kuota Token gaya penulisan sastra

class ParagraphBatch(BaseModel):
    paragraphs: list[str] = Field(description="List of unique paragraphs generated for the topic.")

# =====================================
# Stylized Generation Function
# =====================================

def generate_stylized_batch(topic, batch_size, author):
    # Mengarahkan model meniru kosakata, struktur kalimat, dan atmosfer tulisan sang penulis
    prompt = f"""
    Generate exactly {batch_size} completely distinct and unique paragraphs.
    
    Topic: {topic}
    
    Style Instructions:
    - Write precisely in the unique, distinctive literary style of {author}.
    - Emulate their characteristic sentence structures, choice of vocabulary, rhythmic prose, tone, and atmospheric descriptions.
    - The content must read as if it were penned directly by {author}, reflecting their era's linguistic style.
    
    Requirements:
    - Each paragraph must be between 100 and 200 words.
    - Every paragraph must discuss ONLY the given topic.
    - Original content.
    - Do NOT mention Sherlock Holmes, Arthur Conan Doyle, H.G. Wells or any fictional characters.
    """

    retries = 5
    for attempt in range(retries):
        try:
            response = client.models.generate_content(
                model=MODEL,
                contents=prompt,
                config=types.GenerateContentConfig(
                    response_mime_type="application/json",
                    response_schema=ParagraphBatch,
                    # Temperature sedikit dinaikkan (0.85) agar model lebih kreatif meniru gaya sastra
                    temperature=0.85 
                )
            )
            parsed_response = response.parsed
            return parsed_response.paragraphs

        except Exception as e:
            error = str(e)
            print(f"\n[Error Encountered]: {error}")

            if "RESOURCE_EXHAUSTED" in error:
                wait = 70 * (attempt + 1)
                print(f"Rate limit hit. Sleeping for {wait} seconds...")
                time.sleep(wait)
            elif "UNAVAILABLE" in error:
                print("Server temporarily busy. Waiting 40s...")
                time.sleep(40)
            else:
                wait = 20 * (attempt + 1)
                print(f"Retrying in {wait}s...")
                time.sleep(wait)

    return None

# =====================================
# Smart Resuming Class 3 Loop
# =====================================

author_suffix = TARGET_AUTHOR.replace(' ', '_')

for topic in topics:
    # Nama file disesuaikan khusus untuk Class 3 agar tidak menimpa data Class 2
    safe_filename = f"dataset_Class3_{topic.replace(' ', '_').replace(',', '')}_{author_suffix}.csv"
    topic_rows = []
    
    # Fitur resume otomatis jika proses terhenti di tengah jalan
    if os.path.exists(safe_filename):
        try:
            existing_df = pd.read_csv(safe_filename)
            topic_rows = existing_df.to_dict(orient="records")
            generated = len(topic_rows)
            
            if generated >= PARAGRAPHS_PER_TOPIC:
                print(f"\n[Complete] Class 3 '{topic}' ({TARGET_AUTHOR} style) already has {generated} paragraphs.")
                continue
            else:
                print(f"\n[Resuming] Class 3 '{topic}' ({TARGET_AUTHOR} style) found with {generated}/{PARAGRAPHS_PER_TOPIC}. Finishing...")
        except Exception as e:
            print(f"\n[Warning] Could not read existing file {safe_filename}. Starting fresh. Error: {e}")
            generated = 0
    else:
        print(f"\nStarting Class 3 on Topic: {topic} [Mimicking: {TARGET_AUTHOR}]")
        generated = 0

    while generated < PARAGRAPHS_PER_TOPIC:
        batch = min(BATCH_SIZE, PARAGRAPHS_PER_TOPIC - generated)
        paragraphs_list = generate_stylized_batch(topic, batch, TARGET_AUTHOR)

        if paragraphs_list is None:
            print(f"Skipping remaining batches for '{topic}' due to persistent API issues.")
            break 

        for para in paragraphs_list:
            topic_rows.append({
                "Topic": topic,
                "Paragraph": para.strip(),
                "Class": f"AI_Stylized_{author_suffix}" # Label pembeda untuk klasifikasi teks Anda
            })

        generated += len(paragraphs_list)
        print(f"Progress: Saved {generated}/{PARAGRAPHS_PER_TOPIC} stylized paragraphs.")

        # Simpan progres secara berkala
        df_topic = pd.DataFrame(topic_rows)
        df_topic.to_csv(safe_filename, index=False, encoding="utf-8")

        print("Pacing request... waiting 35 seconds before next batch.")
        time.sleep(35)

print("\nFinished generating all Class 3 stylized topics successfully!")
```
